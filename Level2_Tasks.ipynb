```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Dataset .csv")   # ðŸ‘ˆ use exact filename (with space if present)

print("âœ… Dataset loaded successfully!")
df.head()
```

```python
# Ratings Distribution
ratings = df['Aggregate rating'].value_counts().sort_index()

print("Ratings Distribution:")
print(ratings)

# Bar chart
plt.figure(figsize=(8,4))
ratings.plot(kind='bar', color='teal')
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
# Cuisine Combination Analysis
cuisines = df['Cuisines'].dropna().astype(str).str.replace('/', ',')
cuisines_exploded = cuisines.str.split(',').explode().str.strip()

# Count top 10 cuisines
top_cuisines = cuisines_exploded.value_counts().head(10)

print("Top 10 Cuisines:")
print(top_cuisines)

# Bar chart
plt.figure(figsize=(8,4))
top_cuisines.plot(kind='bar', color='skyblue')
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.show()
```

```python
# Geographic Distribution of Restaurants
plt.figure(figsize=(8,6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.3)
plt.title("Geographic Distribution of Restaurants")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
# Restaurant Chains
chain_counts = df['Restaurant Name'].value_counts().head(10)

print("Top 10 Restaurant Chains:")
print(chain_counts)

# Bar chart
plt.figure(figsize=(10,4))
chain_counts.plot(kind='bar', color='orange')
plt.title("Top 10 Restaurant Chains")
plt.xlabel("Restaurant")
plt.ylabel("Number of Outlets")
plt.show()
```
